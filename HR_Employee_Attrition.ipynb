# Overtime Hours

```python
overtime=df1.pivot_table(values='EmployeeNumber',index='OverTime',columns='Attrition',aggfunc='count')
overtime
```

```python
ax=overtime.plot.bar(
                color=('royalblue','salmon'),
                figsize=(6,6), 
                fontsize=12 )
plt.title('OverTime distribution by Attrition')
plt.legend( shadow=True,
           title='Legend',
           labels=['Stayed','Left'],
            loc='upper right')
plt.xticks([0, 1], ['No', 'Yes'],rotation=45)
plt.ylabel('Number of employess')

for p in ax.patches:
    percentage = f'{p.get_height() /overtime.sum().values.sum() * 100:.2f}%'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom')
    
plt.show()
```

```python
# Correlation-0.25 between OverTime and Attrition--> more employees worked over time, higher chances to leave the company.
# More employees who worked over time decided to left the company, than those who don't, but the correlation is weak.
# It is not causation, there is only correlation.
#Recommendation: A better time management to not overload with work. 
```
